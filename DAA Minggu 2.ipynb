```python
def getSum(myList):
  sum = 0
  for item in myList:
    sum = sum+item
  return sum
getSum([10,11,12,13])
```

```python
def getKali(myList):
  sum = 1
  for item in myList:
    sum = sum*item
  return sum
getKali([10,11,12,13])
```

```python
def getBagi(myList):
  sum = 10000
  for item in myList:
    sum = sum/item
  return sum
getBagi([10,11,12,13])
```

```python
def getSum(myList):
  sum = 0
  for row in myList:
    for item in row:
      sum += item
  return sum
  getSum([[10,11,12],[13,14,15]])
```

```python
def getBagi(myList):
  sum = 100
  for row in myList:
    for item in row:
      sum /= item
  return sum

getBagi([[2],[5]])
```

```python
def getKurang(myList):
  sum = 50
  for row in myList:
    for item in row:
      sum -= item
  return sum
getKurang([[10,2],[5,3]])
```
